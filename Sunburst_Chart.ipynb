# üåû Create Sunburst Chart Using Python

```python
import plotly.graph_objects as go

# Sample data for the Sunburst chart
labels = ['üå± Root', 'üÖ∞Ô∏è Category A', 'üÖ±Ô∏è Category B', 'üÖ∞Ô∏è Category A1', 'üÖ∞Ô∏è Category A2', 'üÖ±Ô∏è Category B1', 'üÖ±Ô∏è Category B2']
parents = ['', 'üå± Root', 'üå± Root', 'üÖ∞Ô∏è Category A', 'üÖ∞Ô∏è Category A', 'üÖ±Ô∏è Category B', 'üÖ±Ô∏è Category B']
values = [100, 30, 70, 10, 20, 40, 30]

# Create the Sunburst chart
fig = go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=values,
    marker=dict(
        colors=values,  # Color the segments based on values
        colorscale='YlGnBu'  # Color scale for the segments
    )
))

# Add title and labels
fig.update_layout(
    title='üåû Sunburst Chart Example',
    uniformtext_mode='hide',  # Hide text if it is too small
)

# Show the plot
fig.show()
```
