# Create Bar Chart Using Python üìä

```python
import matplotlib.pyplot as plt

# Sample data for the bar chart üìä
categories = ['A', 'B', 'Cponding to categories', 'D', 'E']  # Categories
values = [5, 7, 3, 8, 6]               # Values corres

# Create the bar chart üìä
plt.figure(figsize=(8, 5))  # Set the figure size
plt.bar(categories, values, color='skyblue', edgecolor='black', linewidth=1.2)

# Add labels and title ‚úèÔ∏è
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")

# Add grid lines üßÆ
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot üìà
plt.show()
```
