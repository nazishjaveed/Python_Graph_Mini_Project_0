# üå™Ô∏è Create Tornado Chart Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Replace with your dataset
categories = ['Factor A', 'Factor B', 'Factor C', 'Factor D', 'Factor E']
positive_impact = [40, 70, 90, 60, 80]  # Positive impacts (values to the right)
negative_impact = [-30, -60, -80, -50, -70]  # Negative impacts (values to the left)

# Creating the Tornado Chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting positive impacts
ax.barh(categories, positive_impact, color='green', label='‚úÖ Positive Impact')

# Plotting negative impacts (bars to the left)
ax.barh(categories, negative_impact, color='red', label='‚ùå Negative Impact')

# Add labels and title
ax.set_xlabel('Impact Value', fontsize=12)
ax.set_title('üå™Ô∏è Tornado Chart Example', fontsize=16, fontweight='bold')

# Show the legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
