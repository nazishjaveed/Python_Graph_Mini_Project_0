# Create Circular Packet Chart Using Python ðŸ”µ

```python
import matplotlib.pyplot as plt
import circlify

# Data for the chart
data = [
    {'id': 'A', 'value': 50},
    {'id': 'B', 'value': 30},
    {'id': 'C', 'value': 20},
    {'id': 'D', 'value': 10},
    {'id': 'E', 'value': 5}
]

# Prepare data for circlify
circles = circlify.circlify(
    [item['value'] for item in data],
    show_enclosure=False,
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)

# Create figure
fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('off')

# Draw the circles
for circle, item in zip(circles, data):
    x, y, r = circle.x, circle.y, circle.r
    ax.add_patch(plt.Circle((x, y), r, alpha=0.5, linewidth=2, edgecolor='black', facecolor='skyblue'))
    ax.text(x, y, item['id'], size=12, ha='center', va='center')

# Set limits and aspect
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_aspect('equal')

# Show the chart
plt.title('Circular Packing Chart ðŸ”µ', fontsize=16)
plt.show()
```
