# ðŸ§© Create The Mosaic Plot Using Python

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Sample data
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
        'Education': ['Graduate', 'Not Graduate', 'Graduate', 'Graduate', 'Not Graduate', 'Not Graduate']}

df = pd.DataFrame(data)

# Define a function to label the tiles
def labelizer(key):
    return f'{key[0]} ðŸš»\n{key[1]} ðŸŽ“'

# Create the mosaic plot
plt.figure(figsize=(10, 6))
mosaic(df, ['Gender', 'Education'], labelizer=labelizer)
plt.title('Mosaic Plot of Gender ðŸš¹ðŸšº vs. Education ðŸŽ“', fontsize=16)
plt.tight_layout()
plt.show()
```
