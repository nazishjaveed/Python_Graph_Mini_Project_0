
# üç≠ Create Lollipop Chart Using Python


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [5, 8, 12, 7, 10]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a lollipop chart
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

# Plot a scatter plot for the points and a line plot for the lines
plt.scatter(df['Value'], df['Category'], color='blue', s=100, label="Value üîµ", zorder=5)
plt.plot(df['Value'], df['Category'], color='gray', linewidth=2, zorder=1)

# Add labels and title
plt.xlabel('Value üìè', fontsize=12)
plt.ylabel('Category üè∑Ô∏è', fontsize=12)
plt.title('Lollipop Chart üç≠', fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```
