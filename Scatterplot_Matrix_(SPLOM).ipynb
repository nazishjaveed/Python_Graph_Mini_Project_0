# 🔳 Create Scatter Matrix Using Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    '📊 Variable1': [1, 2, 3, 4, 5],
    '📈 Variable2': [2, 4, 1, 8, 7],
    '📉 Variable3': [5, 3, 8, 1, 2],
    '🔢 Variable4': [9, 6, 3, 7, 4],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a scatterplot matrix
sns.pairplot(df, diag_kind='kde', markers='o', palette='Set2')

# Add a title with emoji
plt.suptitle('🌟 Scatterplot Matrix (SPLOM) 🌐', y=1.02, fontsize=16)

# Show the plot
plt.show()
```
