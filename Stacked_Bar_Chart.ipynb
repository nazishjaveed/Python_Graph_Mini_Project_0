# ðŸ“Š Create Stacked Bar Chart Using Python

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data: Replace with your dataset
data = {
    'ðŸ“˜ Category A': [5, 7, 10, 4],
    'ðŸ“— Category B': [3, 2, 4, 8],
    'ðŸ“™ Category C': [6, 9, 7, 3]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['ðŸ”¹ Item 1', 'ðŸ”¸ Item 2', 'ðŸ”¹ Item 3', 'ðŸ”¸ Item 4'])

# Stacked Bar Chart
df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Add titles and labels with emojis
plt.title('ðŸ“Š Stacked Bar Chart Example', fontsize=16, fontweight='bold')
plt.xlabel('ðŸ§¾ Items', fontsize=12)
plt.ylabel('ðŸ“ˆ Values', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
