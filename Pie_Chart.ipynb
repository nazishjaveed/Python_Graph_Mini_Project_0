# 🥧 Create Pie Chart Using Python

```python
import matplotlib.pyplot as plt

# 📊 Sample data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]  # 📈 Percentage distribution
colors = ['gold', 'lightblue', 'lightgreen', 'coral']  # 🎨 Custom colors
explode = (0.1, 0, 0, 0)  # 💥 Emphasize the first slice

# 🧁 Create the pie chart
plt.figure(figsize=(8, 8))  # 📐 Set the figure size
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)

# 📝 Add a title
plt.title("Pie Chart Example 🥧")

# 👀 Show the plot
plt.show()
```
