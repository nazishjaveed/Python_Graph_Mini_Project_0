# ðŸ¥§ Create Pie Chart Using Python

```python
import matplotlib.pyplot as plt

# ðŸ“Š Sample data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]  # ðŸ“ˆ Percentage distribution
colors = ['gold', 'lightblue', 'lightgreen', 'coral']  # ðŸŽ¨ Custom colors
explode = (0.1, 0, 0, 0)  # ðŸ’¥ Emphasize the first slice

# ðŸ§ Create the pie chart
plt.figure(figsize=(8, 8))  # ðŸ“ Set the figure size
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)

# ðŸ“ Add a title
plt.title("Pie Chart Example ðŸ¥§")

# ðŸ‘€ Show the plot
plt.show()
```
