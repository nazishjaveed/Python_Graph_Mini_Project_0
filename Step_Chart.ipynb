# ‚èπÔ∏è Create The Step Chart Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['üÖ∞Ô∏è A', 'üÖ±Ô∏è B', 'üåä C', 'üìà D', '‚≠ê E']
values = [3, 7, 5, 9, 6]  # Values corresponding to categories

# Convert categories to numerical indices for the x-axis
x_positions = np.arange(len(categories))

# Create the step chart
plt.figure(figsize=(8, 5))
plt.step(x_positions, values, where='mid', color='teal', linewidth=2, label='üìä Value')

# Plot markers on the steps
plt.scatter(x_positions, values, color='üî¥', zorder=5, label='üìç Data Points')

# Customize the chart
plt.xticks(x_positions, categories, fontsize=10)  # Set category labels
plt.xlabel('üè∑Ô∏è Category', fontsize=12)
plt.ylabel('üî¢ Values', fontsize=12)
plt.title('üìâ Step Chart', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```
