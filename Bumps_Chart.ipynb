# üìà Create Bumps Chart Using Python üêç

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# üìä Example data for Bump Chart
data = {
    'Year': [2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022],
    'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Rank': [1, 2, 3, 2, 1, 3, 3, 1, 2]
}
df = pd.DataFrame(data)

# üîÑ Pivot the data for plotting
pivot_df = df.pivot(index='Category', columns='Year', values='Rank')

# üé® Create the Bump Chart
plt.figure(figsize=(10, 6))

# üåà Colors for each category
colors = ['#007ACC', '#FF5733', '#50C878']

for idx, (category, row) in enumerate(pivot_df.iterrows()):
    plt.plot(row.index, row.values, marker='o', label=category, color=colors[idx])
    for year, rank in row.items():
        plt.text(year, rank, str(rank), color=colors[idx], fontsize=10, ha='center', va='center')

# üè∑Ô∏è Add labels and grid
plt.gca().invert_yaxis()  # üîÑ Invert y-axis to show ranks properly
plt.xticks(pivot_df.columns, pivot_df.columns, fontsize=10)
plt.yticks(range(1, len(pivot_df) + 2), fontsize=10)
plt.title('üìà Bump Chart', fontsize=14)
plt.xlabel('üìÖ Year', fontsize=12)
plt.ylabel('üèÜ Rank', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='üìÇ Category', fontsize=10)

# üëÄ Show the plot
plt.tight_layout()
plt.show()
```
