# 📉 Create The Rolling Average Plot Using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 📆 Generate Sample Data
np.random.seed(42)
date_range = pd.date_range(start='2025-01-01', periods=100, freq='D')
data = {
    'Date': date_range,
    'Value': np.random.randn(100).cumsum() + 50  # 📈 Random walk data
}
df = pd.DataFrame(data)

# 🔁 Calculate Rolling Average
rolling_window = 7  # 🗓️ Window size in days
df['Rolling_Avg'] = df['Value'].rolling(window=rolling_window).mean()

# 🎨 Plot the Data and Rolling Average
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Value'], label='Original Data 📘', color='blue', alpha=0.6)
plt.plot(df['Date'], df['Rolling_Avg'], label=f'{rolling_window}-Day Rolling Average 🔴', color='red', linewidth=2)

# 🏷️ Titles and Labels
plt.title('Rolling Average Plot 📊🔁', fontsize=16, fontweight='bold')
plt.xlabel('Date 📅', fontsize=12)
plt.ylabel('Value 💹', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.tight_layout()

# ✅ Show the plot
plt.show()
```
