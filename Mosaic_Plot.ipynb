```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import pandas as pd

# Sample Data: Replace with your dataset
data = {
    'Category1': ['A', 'A', 'B', 'B', 'C', 'C', 'C'],
    'Category2': ['X', 'Y', 'X', 'Y', 'X', 'Y', 'X'],
}

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Create a crosstab for the categories
crosstab = pd.crosstab(df['Category1'], df['Category2'])

# Plot the Mosaic Plot
fig, ax = plt.subplots(figsize=(8, 6))
mosaic(crosstab.stack(), ax=ax)
ax.set_title('Mosaic Plot', fontsize=14, fontweight='bold')
plt.show()
```
