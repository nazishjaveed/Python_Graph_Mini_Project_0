# ðŸ“Š Create The Pareto Chart Using Python

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data: Replace with your dataset
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Frequency': [50, 30, 20, 15, 10]}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the data by frequency in descending order
df = df.sort_values(by='Frequency', ascending=False)

# Calculate cumulative percentage
df['Cumulative Percentage'] = df['Frequency'].cumsum() / df['Frequency'].sum() * 100

# Create the figure and bar chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for frequencies
ax1.bar(df['Category'], df['Frequency'], color='skyblue', label='Frequency')
ax1.set_xlabel('Category')
ax1.set_ylabel('Frequency')
ax1.set_title('Pareto Chart')

# Line chart for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(df['Category'], df['Cumulative Percentage'], color='red', marker='o', label='Cumulative %')
ax2.set_ylabel('Cumulative Percentage (%)')
ax2.axhline(80, color='gray', linestyle='--', linewidth=1, label='80% Line')

# Add legends and grid
ax1.legend(loc='upper left')
ax2.legend(loc='lower right')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
