# Create Box Plot Using Python üìä

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the box plot üìä
data = np.random.randn(100)  # Generate 100 random data points (normal distribution)

# Create the box plot üìà
plt.figure(figsize=(8, 6))  # Set the figure size
plt.boxplot(data, vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='blue', linewidth=1.5),
            capprops=dict(color='red', linewidth=2),
            flierprops=dict(markerfacecolor='red', marker='o', markersize=8))

# Add labels and title ‚úèÔ∏è
plt.xlabel("Values")
plt.title("Box Plot Example")

# Show the plot üëÄ
plt.show()
```
