# Create Bubble Chart Using Python 🟢📊

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the bubble chart 🟢📊
np.random.seed(10)
x = np.random.rand(50) * 100  # X values (random values scaled to 100)
y = np.random.rand(50) * 100  # Y values (random values scaled to 100)
sizes = np.random.rand(50) * 1000  # Bubble sizes (random values scaled to 1000)
colors = np.random.rand(50)  # Bubble colors (random values)

# Create the bubble chart 🧑‍💻
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis', edgecolors="w", linewidth=0.5)

# Add labels and title ✏️
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bubble Chart Example")

# Show the plot 👀
plt.show()
```
