```python
import folium
from folium.plugins import HeatMap
import numpy as np

# Create a base map centered on a specific location (e.g., New York)
map_center = [40.7128, -74.0060]  # Latitude, Longitude for New York City
mymap = folium.Map(location=map_center, zoom_start=12)

# Sample data for heatmap (latitude, longitude, and intensity)
# This is just an example; replace with your own geospatial data
locations = [
    [40.7128, -74.0060, 0.5],   # (lat, lon, intensity)
    [40.7328, -74.0060, 0.6],
    [40.7528, -74.0160, 0.7],
    [40.7628, -74.0260, 0.4],
    [40.7828, -74.0360, 0.9],
    [40.7928, -74.0460, 1.0]
]

# Add the heatmap layer to the map
HeatMap(locations).add_to(mymap)

# Save the map as an HTML file
mymap.save("geospatial_heatmap.html")

# Optionally, display the map in Jupyter Notebook
mymap
```
