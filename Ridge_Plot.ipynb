# 🌄 Create The Ridge Plot Using Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 🎲 Generate Sample Data
np.random.seed(42)
data = {
    'Category': np.random.choice(['A 🍎', 'B 🍌', 'C 🍇', 'D 🍉'], size=1000),
    'Value': np.random.normal(loc=0, scale=1, size=1000) * 100  # Simulated data 📈
}

# 🧾 Create a DataFrame
df = pd.DataFrame(data)

# 📏 Set the size of the plot
plt.figure(figsize=(10, 6))

# 🎨 Create the Ridge Plot
sns.kdeplot(
    data=df,
    x='Value',
    hue='Category',
    fill=True,
    common_norm=False,
    palette="crest",
    alpha=0.6,
    linewidth=2
)

# 🏷️ Titles and labels
plt.title('Ridge Plot Example 🌈', fontsize=16, fontweight='bold')
plt.xlabel('Value 📊', fontsize=12)
plt.ylabel('Density 🧪', fontsize=12)

# ✅ Show the plot
plt.tight_layout()
plt.show()
```
