# 🔲 Create Matrix Plot Using Python

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some random data for the matrix (you can replace this with your own data)
data = np.random.rand(10, 10)  # 10x10 matrix of random values

# Create a heatmap (matrix plot)
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)

# Add title and labels
plt.title('Matrix Plot (Heatmap) 🔥🧊', fontsize=16)
plt.xlabel('X Axis ➡️', fontsize=12)
plt.ylabel('Y Axis ⬇️', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
