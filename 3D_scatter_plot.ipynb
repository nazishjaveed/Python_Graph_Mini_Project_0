# Create 3D Scatter Plot Using Python üìäüêç

```python
import matplotlib.pyplot as plt  # üìä Plotting library
from mpl_toolkits.mplot3d import Axes3D  # üõ†Ô∏è 3D plotting toolkit
import numpy as np  # üî¢ Numerical operations

# üé≤ Generate random data for the 3D scatter plot
np.random.seed(42)  # üîÑ For reproducibility
x = np.random.rand(50) * 10  # ‚û°Ô∏è Random x values between 0 and 10
y = np.random.rand(50) * 10  # ‚¨ÜÔ∏è Random y values between 0 and 10
z = np.random.rand(50) * 10  # üîΩ Random z values between 0 and 10
colors = np.random.rand(50)  # üåà Random colors for the points

# üñºÔ∏è Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))  # üìè Set figure size
ax = fig.add_subplot(111, projection='3d')  # üé• 3D projection

# üîµ Plotting the scatter
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis', s=50)

# üé® Add color bar for reference
color_bar = plt.colorbar(scatter, ax=ax)
color_bar.set_label("Color Intensity üåà")

# üè∑Ô∏è Set labels and title
ax.set_xlabel("X Axis ‚û°Ô∏è")
ax.set_ylabel("Y Axis ‚¨ÜÔ∏è")
ax.set_zlabel("Z Axis üîΩ")
ax.set_title("3D Scatter Plot üìä")

# üëÄ Show the plot
plt.show()
```
