# Create ECDF Plot Using Python ðŸ“ˆ

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data: Replace this with your own data
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution

# Plot ECDF using seaborn
sns.ecdfplot(data, label="ECDF", color="blue")

# Customizing the plot
plt.title("Empirical Cumulative Distribution Function (ECDF)", fontsize=16)
plt.xlabel("Data values", fontsize=12)
plt.ylabel("Cumulative Probability", fontsize=12)
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
