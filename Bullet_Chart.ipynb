
# üìä Create Bullet Chart Using Python üêç


```python
import matplotlib.pyplot as plt

# üìä Data for the bullet chart
performance = 70  # üéØ Actual performance
target = 85       # üéØ Target value
ranges = [30, 60, 90]  # üìè Ranges for benchmarks (e.g., low, medium, high)

# üõ†Ô∏è Create the bullet chart
plt.figure(figsize=(8, 2))

# ‚ûñ Plot the benchmark ranges as horizontal bars
plt.barh(0, ranges[2], color='lightgray', height=0.5, label='High üöÄ')
plt.barh(0, ranges[1], color='gray', height=0.5, label='Medium ‚öôÔ∏è')
plt.barh(0, ranges[0], color='darkgray', height=0.5, label='Low üê¢')

# üîµ Plot the actual performance
plt.barh(0, performance, color='blue', height=0.3, label='Performance üíº')

# üî¥ Plot the target as a vertical line
plt.axvline(target, color='red', linestyle='--', linewidth=2, label='Target üéØ')

# üé® Customize the chart
plt.title('üìà Bullet Chart', fontsize=14)
plt.yticks([])  # ‚ùå Remove y-axis ticks
plt.xticks(range(0, 101, 10))  # üìè X-axis ticks
plt.xlabel('Value üìä', fontsize=12)
plt.legend(loc='upper right', fontsize=10)

# ‚ûï Add gridlines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# üëÄ Show the chart
plt.tight_layout()
plt.show()
```
