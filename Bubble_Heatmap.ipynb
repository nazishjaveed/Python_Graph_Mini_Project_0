# Create Bubble Heatmap Using Python üü¢üå°Ô∏èüìä

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the bubble heatmap üü¢üå°Ô∏è
# For example, we generate a 10x10 matrix with values ranging from 1 to 100 for the bubble sizes
data = np.random.rand(10, 10) * 100  # 10x10 matrix with random values scaled for bubble size

# Create a meshgrid for positioning the bubbles üåç
x, y = np.meshgrid(np.arange(data.shape[1]), np.arange(data.shape[0]))

# Flatten the arrays for scatter plot üîÑ
x = x.flatten()
y = y.flatten()
z = data.flatten()  # Bubble size based on data

# Create the plot üé®
plt.figure(figsize=(10, 8))

# Create a scatter plot where the size of each point is determined by the data values üìä
scatter = plt.scatter(x, y, s=z, c=z, cmap='coolwarm', alpha=0.6, edgecolors="w", linewidth=0.5)

# Add color bar to represent the bubble sizes/values üé®
plt.colorbar(scatter, label='Value')

# Add title and labels ‚úèÔ∏è
plt.title("Bubble Heatmap", fontsize=16)
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)

# Set gridlines üó∫Ô∏è
plt.grid(True, which='both', axis='both', color='gray', linestyle='--', linewidth=0.5)

# Show the plot üëÄ
plt.tight_layout()
plt.show()
```
