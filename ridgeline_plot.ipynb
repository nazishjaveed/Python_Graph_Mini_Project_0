# 🌄 Create The Ridgeline Plot Using Python

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 🎯 Create sample data
categories = ['A 🅰️', 'B 🅱️', 'C 🌟', 'D 🌈', 'E 🚀']
data = []

# 🎲 Generate random data for each category
np.random.seed(42)  # For reproducibility 🧬
for i, category in enumerate(categories):
    values = np.random.normal(loc=i, scale=1, size=200)  # Shift mean for each category 🔄
    data.extend(zip([category] * 200, values))

# 🧾 Convert to DataFrame
df = pd.DataFrame(data, columns=['Category', 'Value'])

# 🎨 Create the ridgeline plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# ⛰️ Use seaborn's kdeplot for ridgeline effect
for i, category in enumerate(categories):
    subset = df[df['Category'] == category]
    sns.kdeplot(
        subset['Value'],
        bw_adjust=0.8,
        fill=True,
        alpha=0.6,
        linewidth=1.5,
        label=category,
        clip=(-3, len(categories) + 2),  # Keep within bounds 🔒
        zorder=-i  # Order layers for clarity 📚
    )

# ✨ Customize the plot
plt.title('Ridgeline Plot 🎢', fontsize=16)
plt.xlabel('Value 📊', fontsize=12)
plt.ylabel('Density 🧪', fontsize=12)
plt.legend(title='Category 🔠', fontsize=10, loc='upper left')
plt.tight_layout()

# ✅ Show the plot
plt.show()
```
