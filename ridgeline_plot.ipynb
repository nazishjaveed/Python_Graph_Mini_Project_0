# ðŸŒ„ Create The Ridgeline Plot Using Python

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# ðŸŽ¯ Create sample data
categories = ['A ðŸ…°ï¸', 'B ðŸ…±ï¸', 'C ðŸŒŸ', 'D ðŸŒˆ', 'E ðŸš€']
data = []

# ðŸŽ² Generate random data for each category
np.random.seed(42)  # For reproducibility ðŸ§¬
for i, category in enumerate(categories):
    values = np.random.normal(loc=i, scale=1, size=200)  # Shift mean for each category ðŸ”„
    data.extend(zip([category] * 200, values))

# ðŸ§¾ Convert to DataFrame
df = pd.DataFrame(data, columns=['Category', 'Value'])

# ðŸŽ¨ Create the ridgeline plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# â›°ï¸ Use seaborn's kdeplot for ridgeline effect
for i, category in enumerate(categories):
    subset = df[df['Category'] == category]
    sns.kdeplot(
        subset['Value'],
        bw_adjust=0.8,
        fill=True,
        alpha=0.6,
        linewidth=1.5,
        label=category,
        clip=(-3, len(categories) + 2),  # Keep within bounds ðŸ”’
        zorder=-i  # Order layers for clarity ðŸ“š
    )

# âœ¨ Customize the plot
plt.title('Ridgeline Plot ðŸŽ¢', fontsize=16)
plt.xlabel('Value ðŸ“Š', fontsize=12)
plt.ylabel('Density ðŸ§ª', fontsize=12)
plt.legend(title='Category ðŸ” ', fontsize=10, loc='upper left')
plt.tight_layout()

# âœ… Show the plot
plt.show()
```
