#  Create Decision Tree Visualization Using Python ðŸŒ³ðŸ“Š

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load the Iris dataset (you can replace this with any dataset) ðŸŒ¸ðŸ“Š
data = load_iris()
X = data.data
y = data.target

# Create a Decision Tree classifier ðŸŒ³
clf = DecisionTreeClassifier(random_state=42)

# Fit the model ðŸ”„
clf.fit(X, y)

# Create the plot ðŸŽ¨
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=data.feature_names, class_names=data.target_names, rounded=True)

# Add title with emojis ðŸŒ³ðŸ“Š
plt.title('Decision Tree Visualization ðŸŒ³ðŸ“Š', fontsize=16)

# Show the plot ðŸ‘€
plt.show()
```
