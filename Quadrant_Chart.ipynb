# üìç Create Quadrant Chart Using Python

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# üì¶ Sample data
data = {
    "Item": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "X": [2, 4, 6, 8, 5, 7, 3, 9, 1, 4],
    "Y": [7, 8, 6, 4, 5, 3, 9, 2, 8, 6]
}
df = pd.DataFrame(data)

# ‚ûó Define the midpoint for the quadrants
x_mid = np.mean(df["X"])
y_mid = np.mean(df["Y"])

# üìä Plot
fig, ax = plt.subplots(figsize=(8, 8))

# üîµ Scatter plot
scatter = ax.scatter(df["X"], df["Y"], s=100, c='blue', alpha=0.7)

# üè∑Ô∏è Annotate each point
for i, row in df.iterrows():
    ax.text(row["X"] + 0.2, row["Y"] + 0.2, row["Item"], fontsize=10)

# ‚ûï Draw quadrant lines
ax.axvline(x=x_mid, color='red', linestyle='--', linewidth=1)  # üî¥
ax.axhline(y=y_mid, color='red', linestyle='--', linewidth=1)  # üî¥

# üõ†Ô∏è Customize the chart
ax.set_title("Quadrant Chart üß≠", fontsize=16)
ax.set_xlabel("X Metric ‚û°Ô∏è", fontsize=12)
ax.set_ylabel("Y Metric ‚¨ÜÔ∏è", fontsize=12)
ax.grid(alpha=0.3)

# üß≠ Label quadrants
ax.text(x_mid + 0.5, y_mid + 0.5, "Q1 üí°", fontsize=14, color='green')
ax.text(x_mid - 1.5, y_mid + 0.5, "Q2 üöÄ", fontsize=14, color='green')
ax.text(x_mid - 1.5, y_mid - 0.5, "Q3 üõ†Ô∏è", fontsize=14, color='green')
ax.text(x_mid + 0.5, y_mid - 0.5, "Q4 üìâ", fontsize=14, color='green')

# üì∫ Show the plot
plt.tight_layout()
plt.show()
```
