#  ⚫ Create Scatter Plot Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the scatter plot
x = np.random.rand(50) * 10  # Random x values between 0 and 10
y = np.random.rand(50) * 10  # Random y values between 0 and 10
colors = np.random.rand(50)  # Random colors for the points
sizes = np.random.randint(50, 200, size=50)  # Random sizes for the points

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
scatter = plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7, edgecolors="w", linewidth=0.5)

# Add labels and title with emojis
plt.xlabel("📏 X Axis")
plt.ylabel("📐 Y Axis")
plt.title("🎯 Scatter Plot Example")

# Add color bar for reference
plt.colorbar(scatter, label="🌈 Color Intensity")

# Show the plot
plt.show()
```
