# üßá Create Waffle Chart Using Python

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt

# Example data
data = {
    "üçè Category A": 30,
    "üçä Category B": 20,
    "üîµ Category C": 50
}

# Create the Waffle Chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,  # Number of rows in the waffle chart
    values=data,
    colors=["#4CAF50", "#FFC107", "#2196F3"],  # Colors for the categories
    title={"label": "üßá Waffle Chart Example", "loc": "center", "fontsize": 18},
    labels=[f"{k} ({v}%)" for k, v in data.items()],
    legend={"loc": "upper left", "bbox_to_anchor": (1, 1)}
)

# Display the chart
plt.show()
```
