# ðŸŽ» Create Violin Plot Using Python

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data for the violin plot
np.random.seed(10)
data = np.random.randn(100)  # Generate 100 random data points from a normal distribution

# Create the violin plot
plt.figure(figsize=(8, 6))  # Set the figure size
sns.violinplot(data=data, color='skyblue', inner='quart')

# Add labels and title with emojis
plt.xlabel("ðŸŽ¯ Category")
plt.ylabel("ðŸ“ˆ Values")
plt.title("ðŸŽ» Violin Plot Example")

# Show the plot
plt.show()
```
