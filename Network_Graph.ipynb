# üåê Create The Network Graph Using Python

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple graph object
G = nx.Graph()

# Add nodes with emojis
G.add_node("A üòÄ")
G.add_node("B üöÄ")
G.add_node("C üåü")
G.add_node("D üß†")

# Add edges (connections between nodes)
G.add_edge("A üòÄ", "B üöÄ")
G.add_edge("A üòÄ", "C üåü")
G.add_edge("B üöÄ", "D üß†")
G.add_edge("C üåü", "D üß†")

# Create the plot
plt.figure(figsize=(8, 6))

# Draw the graph with various customization options
nx.draw(G, with_labels=True, node_color='skyblue', node_size=2000, 
        edge_color='gray', font_size=14, font_weight='bold', width=2)

# Set a title
plt.title("Network Graph with Emojis", fontsize=16)

# Show the plot
plt.show()
```
