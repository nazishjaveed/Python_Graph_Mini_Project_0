# ðŸ“Š Create Multivariate Scatter Plot Using Python

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'iris' dataset from seaborn (you can replace this with your dataset)
iris = sns.load_dataset('iris')

# Create a multivariate scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='sepal_length', y='sepal_width',
    hue='species', size='petal_length',
    data=iris, palette='Set1', sizes=(50, 200)
)

# Add title and labels with emojis
plt.title('ðŸŒ¸ Multivariate Scatter Plot: Sepal Length vs Sepal Width ðŸ“', fontsize=16)
plt.xlabel('Sepal Length (cm) ðŸ“', fontsize=12)
plt.ylabel('Sepal Width (cm) ðŸ“', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
