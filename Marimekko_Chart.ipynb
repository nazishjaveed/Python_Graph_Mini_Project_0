# üìä Create Marimekko Chart Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1 üßÆ', 'Category 2 üìà']
subcategories = ['Subcategory A üîπ', 'Subcategory B üî∏']
values = np.array([[0.2, 0.15], [0.3, 0.35]])  # Rows are categories, columns are subcategories

# Calculate widths and heights
category_totals = values.sum(axis=1)
subcategory_ratios = values / category_totals[:, None]
category_ratios = category_totals / category_totals.sum()

# Create the Marimekko chart
fig, ax = plt.subplots(figsize=(8, 6))
x_start = 0
for i, (cat_ratio, sub_ratios) in enumerate(zip(category_ratios, subcategory_ratios)):
    y_start = 0
    for j, sub_ratio in enumerate(sub_ratios):
        ax.bar(
            x=x_start, 
            height=sub_ratio, 
            width=cat_ratio, 
            bottom=y_start, 
            label=f"{categories[i]} - {subcategories[j]}" if y_start == 0 else ""
        )
        y_start += sub_ratio
    x_start += cat_ratio

# Customize the chart
ax.set_title('Marimekko Chart üìä', fontsize=16)
ax.set_xlabel('Categories üè∑Ô∏è', fontsize=12)
ax.set_ylabel('Proportion üìê', fontsize=12)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```
