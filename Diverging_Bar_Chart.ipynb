#  Create Diverging Bar Chart Using Python ðŸ“Š

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data: Replace with your own dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Value': [-30, -20, -10, 10, 20, 30]  # Diverging values
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort data for better visualization
df = df.sort_values('Value').reset_index(drop=True)

# Assign colors based on the sign of the values
df['Color'] = df['Value'].apply(lambda x: 'red' if x < 0 else 'green')

# Plot the Diverging Bar Chart
plt.figure(figsize=(8, 6))
bars = plt.barh(df['Category'], df['Value'], color=df['Color'])

# Add data labels
for bar in bars:
    width = bar.get_width()
    label_x_pos = width + (1 if width > 0 else -1)  # Adjust label position
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2,
             f'{width}', va='center', ha='center' if width < 0 else 'left')

# Styling
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Baseline at 0
plt.title('Diverging Bar Chart', fontsize=14, fontweight='bold')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
