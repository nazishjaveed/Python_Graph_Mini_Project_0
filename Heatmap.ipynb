```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data for the heatmap (random 10x10 matrix)
data = np.random.rand(10, 10)  # Generate a 10x10 matrix with random values between 0 and 1

# Create the heatmap
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(data, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5, linecolor='gray')

# Add title
plt.title("Heatmap Example")

# Show the plot
plt.show()
```
