# 💧 Create Waterfall Chart Using Python

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data with emojis
data = {
    'Category': ['🔰 Start', '💰 Sales', '↩️ Returns', '📢 Marketing', '🏁 End'],
    'Values': [1000, 200, -100, -50, 1050]
}
df = pd.DataFrame(data)

# Calculate cumulative values for waterfall effect
df['Cumulative'] = df['Values'].cumsum()
df['Cumulative_Shift'] = df['Cumulative'].shift(1).fillna(0)

# Set bar colors based on value direction
colors = ['#007ACC' if x >= 0 else '#FF5733' for x in df['Values']]

# Create the plot
plt.figure(figsize=(10, 6))
for i, (start, end, color) in enumerate(zip(df['Cumulative_Shift'], df['Cumulative'], colors)):
    plt.bar(i, end - start, bottom=start, color=color)

# Add labels and grid
plt.xticks(range(len(df)), df['Category'])
plt.title('🪜 Waterfall Chart with Emojis')
plt.ylabel('💸 Values')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
