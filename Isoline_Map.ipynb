# üó∫Ô∏è Create Isoline Map Using Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data (e.g., a 2D Gaussian distribution)
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5 * (X**2 + Y**2))  # A 2D Gaussian function for the example üß†

# Create the figure and axis
plt.figure(figsize=(8, 6))

# Plot the isolines (contours) of the function
contour = plt.contour(X, Y, Z, levels=10, cmap='viridis')

# Add a colorbar for the contour levels
plt.colorbar(contour, label='Function Value üåà')

# Set title and labels
plt.title("Isoline Map üó∫Ô∏è: 2D Gaussian Function üìà", fontsize=16)
plt.xlabel("X-axis ‚û°Ô∏è", fontsize=12)
plt.ylabel("Y-axis ‚¨ÜÔ∏è", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
