# üåà‚ú® Create The Arc Diagram Using Python üêçüìä


```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import Arc

# Create a graph üó∫Ô∏è
G = nx.Graph()

# Add nodes üîµ
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

# Add edges ‚û°Ô∏è
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')]
G.add_edges_from(edges)

# Map nodes to positions along a line üìè
pos = {node: (i, 0) for i, node in enumerate(nodes)}

# Create a figure üñºÔ∏è
fig, ax = plt.subplots(figsize=(10, 5))

# Draw nodes üî∑
nx.draw_networkx_nodes(G, pos, ax=ax, node_color='skyblue', node_size=500, edgecolors='black')

# Draw edges as arcs üéØ
for edge in G.edges:
    x_start, y_start = pos[edge[0]]
    x_end, y_end = pos[edge[1]]
    mid_x = (x_start + x_end) / 2  # Midpoint of the arc üéØ
    height = abs(x_end - x_start) / 2  # Determine arc height üìè
    arc = Arc((mid_x, y_start), width=abs(x_end - x_start), height=height, angle=0, color='gray', lw=1.5)
    ax.add_patch(arc)

# Draw node labels üè∑Ô∏è
nx.draw_networkx_labels(G, pos, ax=ax, font_size=10, font_color='black')

# Customize plot üé®
ax.set_xlim(-0.5, len(nodes) - 0.5)
ax.set_ylim(-1, 1)
ax.axis('off')
plt.title('Arc Diagram üìà', fontsize=16)

# Show plot üëÄ
plt.show()
```
