# ‚ûñ Create Line Chart Using Python

```python
import matplotlib.pyplot as plt

# Sample data for the line chart
x = [1, 2, 3, 4, 5]  # X-axis values ‚û°Ô∏è
y = [2, 3, 5, 7, 11]  # Y-axis values üî¢

# Create the line chart
plt.figure(figsize=(8, 5))  # Set the figure size üìè
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8, label='Prime Numbers üî∑')

# Add labels and title
plt.xlabel("X Axis (Numbers) ‚û°Ô∏è", fontsize=12)
plt.ylabel("Y Axis (Values) üìä", fontsize=12)
plt.title("Line Chart Example üìà", fontsize=16)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```
