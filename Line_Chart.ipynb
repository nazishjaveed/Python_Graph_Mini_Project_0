# ➖ Create Line Chart Using Python

```python
import matplotlib.pyplot as plt

# Sample data for the line chart
x = [1, 2, 3, 4, 5]  # X-axis values ➡️
y = [2, 3, 5, 7, 11]  # Y-axis values 🔢

# Create the line chart
plt.figure(figsize=(8, 5))  # Set the figure size 📏
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8, label='Prime Numbers 🔷')

# Add labels and title
plt.xlabel("X Axis (Numbers) ➡️", fontsize=12)
plt.ylabel("Y Axis (Values) 📊", fontsize=12)
plt.title("Line Chart Example 📈", fontsize=16)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```
