```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull

# Generate random points
np.random.seed(42)  # For reproducibility
points = np.random.rand(30, 2)  # 30 random points in 2D

# Compute the convex hull
hull = ConvexHull(points)

# Plot the points and the convex hull
plt.figure(figsize=(8, 6))
plt.plot(points[:, 0], points[:, 1], 'o', label="Random Points", markersize=8)
plt.plot(points[hull.vertices, 0], points[hull.vertices, 1], 'r-', lw=2, label="Convex Hull")

# Fill the interior of the convex hull
plt.fill(points[hull.vertices, 0], points[hull.vertices, 1], 'r', alpha=0.2)

# Customize the plot
plt.title("Convex Hull Plot", fontsize=16)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
