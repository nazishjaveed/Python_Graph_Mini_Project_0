# ðŸ“Š Create Error Bars Chart Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Replace with your dataset
x = np.arange(1, 6)  # X values
y = [5, 7, 9, 6, 8]  # Y values
yerr = [1, 0.8, 0.6, 0.9, 0.7]  # Error values (y-error)

# Error Bars Chart
plt.errorbar(x, y, yerr=yerr, fmt='o', color='blue', ecolor='red', capsize=5, capthick=2, elinewidth=2)

# Add titles and labels
plt.title('Error Bars Chart Example', fontsize=16, fontweight='bold')
plt.xlabel('X Axis', fontsize=12)
plt.ylabel('Y Axis', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
