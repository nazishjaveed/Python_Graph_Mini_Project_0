# 👥 Create Population Pyramid Chart Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# 👶👵 Example data
age_groups = ["0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-49", 
              "50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80+"]
males = [500, 600, 700, 800, 900, 1000, 1100, 900, 800, 700, 600, 500, 400, 300, 200, 100, 50]
females = [450, 550, 650, 750, 850, 950, 1050, 850, 750, 650, 550, 450, 350, 250, 150, 100, 75]

# 📊 Create a horizontal bar plot
fig, ax = plt.subplots(figsize=(10, 8))

# 📏 Bar width
bar_width = 0.4

# 📈 Age groups index
index = np.arange(len(age_groups))

# 👨 Plot male data
ax.barh(index, [-m for m in males], bar_width, color='blue', label='Males 👨')

# 👩 Plot female data
ax.barh(index + bar_width, females, bar_width, color='pink', label='Females 👩')

# 🏷️ Labels and titles
ax.set_yticks(index + bar_width / 2)
ax.set_yticklabels(age_groups)
ax.set_xlabel('Population 🧮')
ax.set_title('Population Pyramid 📊')
ax.legend()

# 📏 Add grid for better readability
ax.grid(axis='x', linestyle='--', alpha=0.7)

# 🖼️ Display the plot
plt.tight_layout()
plt.show()
```
