# ðŸ”· Create Hexbin Plot Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the hexbin plot
x = np.random.randn(10000)  # Generate random x values
y = np.random.randn(10000)  # Generate random y values

# Create the hexbin plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hexbin(x, y, gridsize=30, cmap='YlGnBu')  # Create hexbin plot with specified grid size and color map

# Add color bar to show the density
plt.colorbar(label='Density')

# Add labels and title
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Hexbin Plot Example")

# Show the plot
plt.show()
```
