# Create Boxen Plot Using Python üìä

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate Sample Data üìä
np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=1000),
    'Value': np.random.normal(loc=0, scale=1, size=1000) * 100  # Simulated data
}

# Create a DataFrame üßÆ
df = pd.DataFrame(data)

# Boxen Plot üìà
plt.figure(figsize=(10, 6))
sns.boxenplot(x='Category', y='Value', data=df, palette='Set2')

# Add titles and labels üìù
plt.title('Boxen Plot Example', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Show the plot üëÄ
plt.tight_layout()
plt.show()
```
